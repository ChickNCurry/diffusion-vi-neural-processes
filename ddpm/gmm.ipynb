```python
from torch import distributions as dist
import torch
from matplotlib import pyplot as plt
```

```python
num_components = 3
mix = dist.Categorical(torch.ones(num_components,))
comp = dist.Normal(torch.randn(num_components,), torch.rand(num_components,))
gmm = dist.MixtureSameFamily(mix, comp)
print(comp.loc, comp.scale)
```

```python
samples = gmm.sample((10000,))
```

```python
plt.hist(samples, bins=100, density=True)
plt.show()
```

```python
num_components_2d = 3
mix_2d = dist.Categorical(torch.ones(num_components_2d,))
comp_2d = dist.Independent(dist.Normal(torch.randn(num_components_2d, 2), torch.rand(num_components_2d, 2)), 1) 
gmm_2d = dist.MixtureSameFamily(mix_2d, comp_2d)
```

```python
samples_2d = gmm_2d.sample((10000,))
```

```python
plt.scatter(samples_2d[:, 0], samples_2d[:, 1])
```
